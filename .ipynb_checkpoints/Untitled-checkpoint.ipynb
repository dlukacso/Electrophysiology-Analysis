```python
def plot_abf(abf):
    ephys_dir = '/media/foldy_lab/Storage_Analysis/Electrophysiology/ABF'
    fname = f'{ephys_dir}/{abf}.abf'
    
    abf = pyabf.ABF(fname)
    
    fig = plt.figure(figsize=(10,4))
    
    for channel in range(2):
        xvals = np.around(abf.sweepX * 1000, 2)
        for col in range(5):
            abf.setSweep(sweepNumber = col, channel=channel)
            yvals = abf.sweepY
            
            ax = fig.add_axes([0.02+ 0.20*col, 0.58 - 0.55*channel, 0.16, 0.40])
            ax.plot(xvals, yvals)
    
    plt.show()
    plt.close()
    
    return
```
